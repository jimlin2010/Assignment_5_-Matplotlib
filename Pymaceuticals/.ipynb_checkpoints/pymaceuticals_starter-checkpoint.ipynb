```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

from scipy.stats import linregress
from sklearn import datasets


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_result = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
mouse_result.head()
```

```python
mouse_result.to_csv('data/mouse_result.csv')
mouse_result
```

```python
# Checking the number of mice.
mouse_count = mouse_result['Mouse ID'].drop_duplicates().shape[0]
print(mouse_count)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse = mouse_result.loc[mouse_result.duplicated(subset=['Mouse ID', 'Timepoint']), 'Mouse ID'].unique()
duplicate_mouse
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_df = mouse_result.loc[mouse_result['Mouse ID'] == "g989", :]
duplicate_mouse_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_result_clean = mouse_result[mouse_result['Mouse ID'].isin(duplicate_mouse) == False]
mouse_result_clean
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count_clean = mouse_result_clean['Mouse ID'].drop_duplicates().shape[0]
print(mouse_count_clean)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean = mouse_result_clean['Tumor Volume (mm3)'].groupby(mouse_result_clean['Drug Regimen']).mean()
median = mouse_result_clean['Tumor Volume (mm3)'].groupby(mouse_result_clean['Drug Regimen']).median()
var = mouse_result_clean['Tumor Volume (mm3)'].groupby(mouse_result_clean['Drug Regimen']).var()
std = mouse_result_clean['Tumor Volume (mm3)'].groupby(mouse_result_clean['Drug Regimen']).std()
sem = mouse_result_clean['Tumor Volume (mm3)'].groupby(mouse_result_clean['Drug Regimen']).sem()

mouse_summary = pd.DataFrame({"Mean Tumor Volume": mean,
                              "Median Tumor Volume": median,
                              "Tumor Volume Variance": var,
                              "Tumor Volume Std. Dev.": std,
                              "Tumor Volume Std. Err.": sem})
mouse_summary
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.

agg_summary = mouse_result_clean.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std','sem'])
agg_summary
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

Capomulin = mouse_result_clean.loc[mouse_result_clean["Drug Regimen"] == "Capomulin"]

cap_plot = Capomulin.loc[Capomulin["Mouse ID"] == "l509",:]
x_axis = cap_plot["Timepoint"]
y_axis = cap_plot["Tumor Volume (mm3)"]

plt.title('Capomulin treatmeant of mouse l509')
plt.plot(x_axis, y_axis,linewidth=1, color="blue")
plt.xlabel('Timepoint (Days)')
plt.ylabel('Tumor Volume (mm3)')

```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

average_vol =Capomulin.groupby(['Mouse ID']).mean()

plt.scatter(average_vol['Weight (g)'],average_vol['Tumor Volume (mm3)'], color="blue")
plt.xlabel('Weight (g)')
plt.ylabel('Averag Tumor Volume (mm3)')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(average_vol['Weight (g)'],average_vol['Tumor Volume (mm3)'])
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")

(slope, intercept, rvalue, pvalue, stderr) = linregress(average_vol['Weight (g)'], average_vol['Tumor Volume (mm3)'])
regression=average_vol["Weight (g)"]* slope + intercept


plt.scatter(average_vol['Weight (g)'],average_vol['Tumor Volume (mm3)'], color="blue")
plt.plot(average_vol["Weight (g)"], regression, color='red')
plt.xlabel('Weight (g)')
plt.ylabel('Averag Tumor Volume (mm3)')

```
